```python
import pandas as pd

train_df = pd.read_csv("../data/processed/train_balanced.csv")
test_df = pd.read_csv("../data/processed/test_data.csv")

train_df.head()
```

```python
X_train = train_df.drop("loan_status", axis=1)
y_train = train_df["loan_status"]

X_test = test_df.drop("loan_status", axis=1)
y_test = test_df["loan_status"]
```

```python
import pandas as pd

from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier

from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
log_model = LogisticRegression()

log_model.fit(X_train, y_train)

log_pred = log_model.predict(X_test)
```

```python
print("Logistic Regression Accuracy:", accuracy_score(y_test, log_pred))
print(classification_report(y_test, log_pred))
print(confusion_matrix(y_test, log_pred))
```

```python
svm_model = SVC()

svm_model.fit(X_train, y_train)

svm_pred = svm_model.predict(X_test)
```

```python
print("SVM Accuracy:", accuracy_score(y_test, svm_pred))
print(classification_report(y_test, svm_pred))
print(confusion_matrix(y_test, svm_pred))
```

```python
knn_model = KNeighborsClassifier(n_neighbors=5)

knn_model.fit(X_train, y_train)

knn_pred = knn_model.predict(X_test)
```

```python
print("KNN Accuracy:", accuracy_score(y_test, knn_pred))
print(classification_report(y_test, knn_pred))
```

```python
results = pd.DataFrame({
    "Model": ["Logistic Regression", "SVM", "KNN"],
    "Accuracy": [
        accuracy_score(y_test, log_pred),
        accuracy_score(y_test, svm_pred),
        accuracy_score(y_test, knn_pred)
    ]
})

results
```

Model Evaluation (SVM)

The Support Vector Machine (SVM) model achieved an accuracy of approximately
94.26% on the test dataset. The model demonstrates strong performance across
both classes with high precision, recall, and F1-scores.

The confusion matrix indicates that most loan approvals and rejections are
correctly predicted, with only a small number of misclassifications.
Overall, SVM provides a reliable classification model for predicting loan
approval outcomes.

## Hyperparamter Tunning

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
```

```python
param_grid = {
    "C": [0.1, 1, 10, 100],
    "kernel": ["linear", "rbf"],
    "gamma": ["scale", "auto"]
}
```

```python
grid = GridSearchCV(
    SVC(),
    param_grid,
    cv=5,
    scoring="accuracy",
    n_jobs=-1
)
```

```python
grid.fit(X_train, y_train)
```

```python
print("Best Parameters:", grid.best_params_)
```

```python
best_svm = grid.best_estimator_
```

```python
y_pred = best_svm.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report

print("Tuned SVM Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
import pickle

pickle.dump(best_svm, open("../models/svm_model.pkl", "wb"))
```

### After hyperparameter tuning, the Support Vector Machine (SVM) model achieved an accuracy of 95.67% on the test dataset. The model demonstrates strong performance across both classes with balanced precision, recall, and F1-scores. These results indicate that the tuned SVM model effectively captures the relationship between financial attributes and loan approval outcomes.
